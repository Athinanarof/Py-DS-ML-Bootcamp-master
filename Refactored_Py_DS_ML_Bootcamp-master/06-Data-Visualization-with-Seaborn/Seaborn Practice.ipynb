```python
# Matrixes
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
tips_correlation = tips.corr()
tips_correlation
```

```python
sns.heatmap(tips_correlation, annot=True) # cmap='coolwarm'
```

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
flights_pivot
```

```python
sns.heatmap(flights.corr())
```

```python
flights.corr()
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot, linecolor='white', linewidths=1) #cmap='magma'    
```

```python
sns.clustermap(flights_pivot, standard_scale=1)
```

```python
# Grids

import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head ()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# Regression plots
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
#linear model plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':100})
#linear model plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, size=8)
```
