```python
# Categorical Plots

import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x= 'sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y ='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
# Matrixes
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
tips_correlation = tips.corr()
tips_correlation
```

```python
sns.heatmap(tips_correlation, annot=True) # cmap='coolwarm'
```

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
flights_pivot
```

```python
sns.heatmap(flights.corr())
```

```python
flights.corr()
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot, linecolor='white', linewidths=1) #cmap='magma'    
```

```python
sns.clustermap(flights_pivot, standard_scale=1)
```
