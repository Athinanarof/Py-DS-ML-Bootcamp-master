```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x, y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('My first chart - Numbers Squared')
```

```python
plt.subplot(1, 2, 1) # rows, columns, plot number
plt.plot(x, y, 'r-')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b-')
```

```python
 # Object Oriented way
    
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height

axes.plot(x ,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Plot made with OOP')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2  = fig.add_axes([0.15, 0.7, 0.25, 0.25])

axes1.plot(x, y)
axes2.plot(y, x)
```

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot(x, y, 'r-')
axes[0].set_title('Fig 1')

axes[1].plot(y, x)
axes[1].set_title('Fig 2')

plt.tight_layout()
```

```python
# Figure and Size

fig = plt.figure(figsize=(3, 2), dpi=100) #pixes per inch


ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(1, 1) #pixes per inch


axes.plot(x , y)
```

```python
fig, axes = plt.subplots(1, 2) #pixes per inch


axes[0].plot(x, y)
axes[1].plot(y, x, 'r-')
```

```python
fig.savefig('I did a plot with Matplotlib.png', dpi=200)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,1,1])

axes.plot(x, x**2, label='y squared')
axes.plot(x, x**3, label='x squared')

axes.legend(loc=0)
```

```python
# Plot Appearance

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(x, y, color='purple', linewidth=3, linestyle ='-', marker='o', markersize=10, markerfacecolor='yellow') #lw, ls
```
