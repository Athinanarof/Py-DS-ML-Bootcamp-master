```python
from sklearn.model_selection import train_test_split
```

```python
df[['feature1', 'feature2']].head()
```

```python
X = df[['feature1', 'feature2']].values # Becuase of how tensorflow works we have to pass numpy arrays instead pandas dataframes or series
```

```python
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# For NN we need to normalize and scale the data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense # Dense layers is the one when each node is connected to each node of the next layer
```

```python
model = Sequential([Dense(units=4, activation='relu'), # First layer
                    Dense(units=2, activation='relu'),
                    Dense(units=1)]) # Last layer
                   
# units: how many nodes/neurones are going to be in this layer, activation: like sigmod activation function
# relu: rectified linear unit
```

```python
model = Sequential()

model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1)) # Since this is the last one, we don't really want a activation function

model.compile(optimizer='rmsprop', loss='mse')

# optimizer is for the gradient descent
# mse: mean square error
```

# Choosing an optimizer and loss

Keep in mind what kind of problem you are trying to solve:

    # For a multi-class classification problem
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # For a binary classification problem
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # For a mean squared error regression problem
    model.compile(optimizer='rmsprop',
                  loss='mse')

```python
model.fit(x=X_train, y=y_train, epochs=250)

# 1 epoch means you are going through the entire dataset 1 time
```

```python
model.history.history
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
model.evaluate(X_train, y_train, verbose=0)
```
